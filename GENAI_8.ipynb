```python
import torch
from speechbrain.pretrained import Tacotron2, HIFIGAN
import torchaudio
import IPython.display as ipd

# Load Tacotron2 model (Text-to-Mel Spectrogram)
tacotron2 = Tacotron2.from_hparams(source="speechbrain/tts-tacotron2-ljspeech", savedir="tmp_tts")

# Load HiFi-GAN vocoder (Mel Spectrogram to Audio)
hifigan = HIFIGAN.from_hparams(source="speechbrain/tts-hifigan-ljspeech", savedir="tmp_vocoder")

# Input text
text = "Hello, this is a Generative AI test 7."

# Convert text to mel spectrogram
mel_output, mel_length, alignment = tacotron2.encode_text(text)

# Convert mel spectrogram to waveform
waveforms = hifigan.decode_batch(mel_output)

# Save to file
torchaudio.save("output.wav", waveforms.squeeze(1), 22050)
print("Speech synthesis complete! Saved as output.wav")

# Play the audio
ipd.Audio("output.wav")
```
